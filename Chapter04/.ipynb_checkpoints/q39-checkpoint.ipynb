```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sns; sns.set(font="IPAexGothic")
%config InlineBackend.figure_format = 'retina'

import collections


def get_morphemes(f_name='neko.txt.mecab'):
    morphemes = []
    
    with open(f_name) as f:
        for line in f.readlines():
            line = line.replace('\n', '')

            cols = line.split('\t')
            if len(cols) < 2:
                continue

            surface = cols[0]
            cols = cols[1]
            cols = cols.split(',')

            morpheme = {
                'surface': surface,
                'base': cols[6],
                'pos': cols[0],
                'pos1': cols[1] 
            }
            morphemes.append(morpheme)
            
    return morphemes
```

```python
morphemes = get_morphemes()
base_list = [morpheme['base'] for morpheme in morphemes]
c = dict(collections.Counter(base_list).most_common())
```

```python
plt.scatter(list(range(len(c))), c.values())
plt.xscale('log')
plt.yscale('log')
plt.xlim(1, len(ï½ƒ)+1)
plt.title("Zipf's law")
plt.xlabel('Rank')
plt.ylabel('Frequency')
```
